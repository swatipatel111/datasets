#simple linear regression

```python
#step-1 raed dataset
import pandas as pd
df=pd.read_csv("Book1.csv")
print(df)
#step-2 define x and y
x=df.iloc[:,0:1] #df[["cgpa"]]....2d
y=df.iloc[:,-1]  #df["package"]....1d
#step-3 split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                               test_size=0.2,
                                              random_state=40)
#step-4 train model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
#step-5 make prediction for x_test data
y_pred=lr.predict(x_test)
print(x_test,"\n",y_test,"\n",y_pred)
#step-6 evaluate the model
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("MAE:",mean_absolute_error(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
print("r2score:",r2_score(y_test,y_pred))
#step-7 predict package for your cgpa
import numpy as np
x_t=np.array([8.7]).reshape(1,-1)
print(x_t)
predicted=lr.predict(x_t)
print(predicted)

```

```python
#step-8 graphical analysis 
import matplotlib.pyplot as plt
plt.scatter(df["cgpa"],y)
plt.plot(df["cgpa"],lr.predict(x))
plt.show()
```

```python
#step-1 raed dataset
import pandas as pd
df=pd.read_csv("advertising.csv")
print(df)
#step-2 define x and y
x=df.iloc[:,0:1] 
y=df.iloc[:,-1]  
#step-3 split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                               test_size=0.15,
                                              random_state=20)
#step-4 train model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
#step-5 make prediction for x_test data
y_pred=lr.predict(x_test)
#step-6 evaluate the model
from sklearn.metrics import mean_squared_error,r2_score
print("MSE:",mean_squared_error(y_test,y_pred))
print("r2score:",r2_score(y_test,y_pred))
```

```python
#multiple linear regression
import pandas as pd
df=pd.read_csv("car data.csv")
print(df.head())
df["age"]=2026-df["Year"]
print(df.info())
x=df[["age","Present_Price","Kms_Driven","Fuel_Type","Seller_Type","Transmission",
      "Owner"]]
y=df["Selling_Price"]
x=pd.get_dummies(x,drop_first=True)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,
                                              random_state=20)
#step-4 train model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
#step-5 make prediction for x_test data
y_pred=lr.predict(x_test)
#step-6 evaluate the model
from sklearn.metrics import mean_squared_error,r2_score
print("MSE:",mean_squared_error(y_test,y_pred))
print("r2score:",r2_score(y_test,y_pred))
```
