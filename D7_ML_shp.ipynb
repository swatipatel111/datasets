```python
#step-1 raed dataset
import pandas as pd
df=pd.read_csv("Book1.csv")
print(df)
#step-2 define x and y
x=df.iloc[:,0:1] #df[["cgpa"]]....2d
y=df.iloc[:,-1]  #df["package"]....1d
#step-3 split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                               test_size=0.2,
                                              random_state=40)
#step-4 train model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
#step-5 make prediction for x_test data
y_pred=lr.predict(x_test)
print(x_test,"\n",y_test,"\n",y_pred)
#step-6 evaluate the model
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("MAE:",mean_absolute_error(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
print("r2score:",r2_score(y_test,y_pred))
#step-7 predict package for your cgpa
import numpy as np
x_t=np.array([8.7]).reshape(1,-1)
print(x_t)
predicted=lr.predict(x_t)
print(predicted)

```

```python
#step-8 graphical analysis 
import matplotlib.pyplot as plt
plt.scatter(df["cgpa"],y)
plt.plot(df["cgpa"],lr.predict(x))
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("Ice_cream selling data.csv")
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly=PolynomialFeatures(degree=4)
poly_x=poly.fit_transform(x)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(poly_x,y,test_size=0.15,
                                               random_state=25)
lr=LinearRegression()
lr.fit(x_train,y_train)

y_pred=lr.predict(x_test)
print(y_pred)

from sklearn.metrics import mean_squared_error,r2_score
print("mse:",mean_squared_error(y_test,y_pred),"r2score:",r2_score(y_test,y_pred))
```

```python
df=pd.read_csv("diabetes.csv")
print(df)
x=df.iloc[:,0:8]
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.15,
                                               random_state=25)
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier(n_neighbors=27)
model=kn.fit(x_train,y_train)
pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
#find accuracy and error rate
TN=cm[0][0]
TP=cm[1][1]
FN=cm[1][0] #type 1 error
FP=cm[0][1]#type 2 error
accuracy=(TP+TN)/(TP+TN+FP+FN)
print(accuracy)
print("error rate:",1-accuracy)
```
