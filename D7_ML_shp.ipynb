```python
#step-1 load dataset
import pandas as pd
df=pd.read_csv("diabetes.csv")
print(df)
#step-2 define x(input columns) and y(target column)
x=df.iloc[:,0:8]
y=df.iloc[:,-1]
# step 3 split dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#step-4 import model
from sklearn.svm import SVC
#step-5 train model using SVC
svc_model=SVC(C=10,random_state=42,kernel="rbf")
svc_model.fit(x_train,y_train)
y_pred=svc_model.predict(x_test)
#C regulization parameter high value of C will increase margin distance and reduce error..
#it may go in overfitting
#low value of C will decrease marginal distance and reduce accuracy
#kernal will convert dataset from higher dimension to make it saperable
#rbf radial bias function, most powerfull and commonly used
# step-6 evaluate model
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print("cm:",confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred),":acc")
print(classification_report(y_test,y_pred))
```
