```python
import pandas as pd
df=pd.read_csv("car data.csv")
df.head()
```

```python
import pandas as pd
df=pd.read_csv("diabetes.csv")
df.head()
x=df.iloc[:,0:8]
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.svm import SVC
s=SVC(kernel="rbf",random_state=42,gamma="scale",C=10) 
#C= regulaization parameter high C= high -->more margin& less error
#C=low-->small margin & more error
# gamma-- controls influence of points(for rbf)---high gamma-overfitting,
#low gamma- underfitting
# kernal tranforms data into higher dimension to make it seperable
# kernal= types of kernal (linear, rbf,poly) 
#rbf-most powerfull and commonly used in classification
s.fit(x_train,y_train)
y_pred=s.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
print("cm: \n",confusion_matrix(y_test,y_pred))
print("acc: ",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# 9 marks practice 

import pandas as pd
df=pd.read_csv("Employee_attrition.csv")
df=df.drop("Employee_ID",axis=1)
x=df.drop("Attrition",axis=1)
y=df["Attrition"]
x=pd.get_dummies(x,drop_first=True)
print(x)

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
acc_list=[]
for k in range(1,26):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    acc_list.append(acc)
    print(f"k={k},accuracy={acc}")
    
m=max(acc_list)
best_k=acc_list.index(m)+1
print("best_k=",best_k)

knn_final=KNeighborsClassifier(n_neighbors=best_k)
knn_final.fit(x_train,y_train)
#random forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion="entropy",n_estimators=150,random_state=42)
rf.fit(x_train,y_train)

y_pred1=rf.predict(x_test)
print("for random forest")
print("cm=",confusion_matrix(y_test,y_pred1))
print("accuracy",accuracy_score(y_test,y_pred1))
# prediction for new data
import numpy as np

x_t=np.array([30,45000,5,3,10,True,False,False,True,False,True,False,False]).reshape(1,-1)
print("prediction for knn=",knn_final.predict(x_t))
print("prediction for rf=",rf.predict(x_t))
```

```python
#write a python program to predict booking_status using
#random forest and svc for "Hotel Reservation.csv"
# load dataset, drop bookingt_ID column, convert categorical to numeric
#split data in to train and test(80%-20%),random_state=42
#random forest: 
#train random forest model for values of n_estimators from 61 to 70,criterion="entropy"
#random_state=42
#find the best n_estimators based on highest accuracy
#train the final random forest model using best 

#svc:
#train svc model using C=10,random_state=42,kernel="rbf"
#find accuracy and confusion matrix
```

```python
import pandas as pd
df=pd.read_csv("Hotel Reservations.csv")
print(df.info())
df=df.drop("Booking_ID",axis=1)
x=df.drop("booking_status",axis=1)
y=df["booking_status"]
x=pd.get_dummies(x,drop_first=True)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
acc_list=[]
for e in range(61,71):
    rf=RandomForestClassifier(n_estimators=e,random_state=42,criterion="entropy")
    rf.fit(x_train,y_train)
    y_pred=rf.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    acc_list.append(acc)
    print(f"e={e},accuracy={acc}")  
m=max(acc_list)
best_e=acc_list.index(m)+61
print("best_e=",best_e)
rf_final=RandomForestClassifier(n_estimators=best_e,criterion="entropy",random_state=42)
rf_final.fit(x_train,y_train)
y_pred1=rf_final.predict(x_test)
print("for random forest")
print("cm=",confusion_matrix(y_test,y_pred1))
print("accuracy",accuracy_score(y_test,y_pred1))
#svc
from sklearn.svm import SVC
s=SVC(C=40,random_state=42,kernel="rbf")
s.fit(x_train,y_train)
y_pred1=s.predict(x_test)
print("for svc")
print("cm=",confusion_matrix(y_test,y_pred1))
print("accuracy",accuracy_score(y_test,y_pred1))

```
